```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../vehicles_us.csv")
df.head()
```

```python
# price by vehicle condition
px.box(df, x="condition", y="price", color="condition",
       title="Price Distribution by Condition")
```

```python
# odometer vs price
px.scatter(df, x="odometer", y="price", color="type",
           title="Odometer vs Price by Vehicle Type")
```
